```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series, DataFrame


train_path ="Data/train.csv"
df=pd.read_csv(train_path)
```

```python
#view the data
df
```

```python
#some statistics about the numerical data
df.describe()
```

Age is the only numerical feature with NaNs. <br>

```python
# specifies the parameters of our graphs
fig = plt.figure(figsize=(18,6))
alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.6

# lets us plot many diffrent shaped graphs together 
ax1 = plt.subplot2grid((2,3),(0,0))

# plots a bar graph of those who surived vs those who did not.               
df.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax1.set_xlim(-1, 2)
plt.title("Survival count")

# plots a bar graph of the age distribution
ax3 = plt.subplot2grid((2,3),(0,1))
df.Age.plot(kind='kde',alpha=alpha_bar_chart)
plt.title("Age distribution")


# plots a bar graph of the class distribution
ax3 = plt.subplot2grid((2,3),(0,2))
df.Pclass.value_counts().plot(kind="bar", alpha=alpha_bar_chart)
plt.title("Class Distribution")

plt.show()
```
